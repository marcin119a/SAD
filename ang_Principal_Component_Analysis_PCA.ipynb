### PCA

Algorytm PCA (ang. Principal Component Analysis) to jedna z popularnych metod analizy danych, stosowana w celu redukcji wymiarowości oraz ekstrakcji cech z danych wielowymiarowych.

Algorytm PCA polega na transformacji danych wejściowych, tak aby uzyskać nowe, niewspółliniowe zbiory zmiennych (tzw. główne składowe), które maksymalizują wariancję zbioru danych. Każda kolejna główna składowa jest liniowo nieskorelowana z poprzednimi i odpowiada kierunkowi o największej zmienności w danych.

Dzięki temu algorytmowi można zmniejszyć liczbę wymiarów w danych, zachowując przy tym jak najwięcej informacji. Pozwala to na efektywną wizualizację danych oraz poprawę jakości modelowania. Algorytm PCA znajduje zastosowanie w wielu dziedzinach, takich jak np. analiza obrazów, rozpoznawanie mowy, biometria czy analiza finansowa.

Mamy zbiór danych o wymiarze 1000 x 500, gdzie każda z 1000 próbek ma 500 cech. Chcesz zredukować wymiarowość tego zbioru danych za pomocą algorytmu PCA.

Jakie kroki powinieneś podjąć, aby wyznaczyć nowe zmienne, które będą najlepiej reprezentować oryginalne cechy?

Podpowiedź: przedstaw, jakie kroki są potrzebne do wykonania analizy PCA oraz jakie informacje można otrzymać dzięki tej metodzie.

Oto przykładowe rozwiązanie problemu w języku Python z użyciem biblioteki scikit-learn:


```python

from sklearn.decomposition import PCA
import numpy as np
import pandas as pd

# Załadowanie danych
data = pd.read_csv('https://raw.githubusercontent.com/marcin119a/PODSTAWY-UCZENIA-MASZYNOWEGO-W-PYTHONIE-/main/income.csv', sep=';')
data = data[['GDP per capita', 'Life expectancy']]


# Wykonanie standaryzacji danych
mean = np.mean(data, axis=0)
std = np.std(data, axis=0)
data_standardized = (data - mean) / std

# Tworzenie instancji klasy PCA i wyznaczanie głównych składowych
pca = PCA(n_components=2)  # redukcja do dwóch wymiarów
principal_components = pca.fit_transform(data_standardized)

# Wyświetlenie wyjaśnionej wariancji dla każdej z głównych składowych
print("Wyjaśniona wariancja: ", pca.explained_variance_ratio_)

# Wykres punktowy nowych zmiennych
import matplotlib.pyplot as plt

plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.title("Wizualizacja zbioru danych po redukcji wymiarowości PCA")
plt.xlabel("Główna składowa 1")
plt.ylabel("Główna składowa 2")
plt.show()
```
